<h3>Mise a l'échelle des caractéristiques</h3>

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('nba.csv')
df.head()
```

```python
x = df[['Age','Salary']]
mise_a_echelle=StandardScaler()
x=mise_a_echelle.fit_transform(x)
print(x)
```

```python
dfx=pd.DataFrame(data= x,columns=['Age2','Salary2'])
df=pd.concat([df,dfx],axis=1)
df=df.drop(['Age','Salary'],axis=1)
print(df)
```
